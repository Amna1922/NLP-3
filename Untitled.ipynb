# 0. Environment Setup & Imports

```python
import os
import gzip
import json
import random
import math
import itertools
from collections import Counter
from typing import List, Tuple, Dict, Optional

import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import matplotlib.pyplot as plt
from tqdm import tqdm

# Set seeds for reproducibility
RANDOM_SEED = 42
random.seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)
torch.manual_seed(RANDOM_SEED)
torch.cuda.manual_seed_all(RANDOM_SEED)

# Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Create directories
os.makedirs("models", exist_ok=True)
os.makedirs("results", exist_ok=True)
```

# 1. Dataset Preparation

We load reviews from three categories, sample ~12k per category, and split into train/val/test (70/15/15). The helpful field is used to define a binary “is helpful” target (1 if at least one person voted helpful).

```python
# ---- Load raw data from .json.gz files ----
def load_gz_json(path: str) -> List[dict]:
    with gzip.open(path, 'rt', encoding='utf-8') as f:
        data = [json.loads(line) for line in f]
    return data

paths = {
    "cellphones": "data/cellphones.json.gz",
    "beauty": "data/beauty.json.gz",
    "sports": "data/sports.json.gz"
}
raw = {}
for cat, p in paths.items():
    raw[cat] = load_gz_json(p)
    print(f"{cat}: {len(raw[cat])} reviews loaded")

# ---- Build subset: 12,000 per category (or as many as available) ----
SAMPLES_PER_CATEGORY = 12000
subset = []
for cat, reviews in raw.items():
    if len(reviews) >= SAMPLES_PER_CATEGORY:
        sampled = random.sample(reviews, SAMPLES_PER_CATEGORY)
    else:
        sampled = reviews  # use all if not enough
    print(f"{cat}: selected {len(sampled)} reviews")
    subset.extend(sampled)

# Shuffle combined data
random.shuffle(subset)
print(f"Total reviews in subset: {len(subset)}")

# ---- Train/val/test split ----
n_total = len(subset)
n_train = int(0.7 * n_total)
n_val = int(0.15 * n_total)
n_test = n_total - n_train - n_val

train_data = subset[:n_train]
val_data = subset[n_train:n_train+n_val]
test_data = subset[n_train+n_val:]
print(f"Train: {len(train_data)}, Val: {len(val_data)}, Test: {len(test_data)}")
```

# 2. Preprocessing

We implement a simple tokenizer (word-level) that lowercases, splits on non-alphabetic characters, and builds a vocabulary from the training set only. Special tokens are added for later use.

```python
# ---- Custom Tokenizer ----
class SimpleTokenizer:
    def __init__(self, special_tokens=None, max_vocab_size=10000):
        if special_tokens is None:
            special_tokens = ["<PAD>", "<UNK>", "<BOS>", "<EOS>",
                              "<REVIEW>", "<SENT>", "<HELP>", "<RETR>", "<SEP>", "<EXP>"]
        self.special_tokens = special_tokens
        self.max_vocab = max_vocab_size
        self.word2idx = {}
        self.idx2word = {}
        self.vocab_size = 0

    def tokenize(self, text: str) -> List[str]:
        # Lowercase and split on non-alphabetic (keep spaces for readability)
        text = text.lower()
        # Split on any character that is not a letter or digit, keep apostrophes etc.
        tokens = []
        for word in text.split():
            # Remove punctuation attached to word boundaries but keep internal apostrophes
            word = word.strip(".,;:!?\"'()[]{}<>|/\\*#@&%$+-=~")
            if word:
                tokens.append(word)
        return tokens

    def build_vocab(self, texts: List[str]):
        counter = Counter()
        for text in texts:
            counter.update(self.tokenize(text))
        # Keep most common words up to max_vocab minus special tokens
        common_words = [word for word, _ in counter.most_common(self.max_vocab - len(self.special_tokens))]
        self.word2idx = {word: i for i, word in enumerate(common_words)}
        # Add special tokens at the end (so they do not interfere with word indices)
        for token in self.special_tokens:
            if token not in self.word2idx:
                self.word2idx[token] = len(self.word2idx)
        self.idx2word = {v: k for k, v in self.word2idx.items()}
        self.vocab_size = len(self.word2idx)
        print(f"Vocabulary size: {self.vocab_size}")

    def encode(self, text: str, max_len: int = 512, add_special: bool = False) -> List[int]:
        tokens = self.tokenize(text)
        ids = [self.word2idx.get(token, self.word2idx["<UNK>"]) for token in tokens]
        if add_special:
            ids = [self.word2idx["<BOS>"]] + ids + [self.word2idx["<EOS>"]]
        # Truncate if necessary
        if len(ids) > max_len:
            ids = ids[:max_len]
        return ids

    def decode(self, ids: List[int]) -> str:
        tokens = [self.idx2word.get(i, "<UNK>") for i in ids]
        return " ".join(tokens)

# Prepare training texts for vocab building
train_texts = [item["reviewText"] for item in train_data]
tokenizer = SimpleTokenizer(max_vocab_size=10000)
tokenizer.build_vocab(train_texts)
```

Next, we define the maximum sequence length, pad/truncate all sequences, and create PyTorch Datasets for encoder training. For each sample we also prepare:

Sentiment label: 0 (negative: 1-2), 1 (neutral: 3), 2 (positive: 4-5)

Helpfulness label: 1 if helpful[0] > 0, else 0

```python
MAX_SEQ_LEN = 256   # reduced for faster experiments; you may increase

def get_sentiment_label(rating: float) -> int:
    if rating <= 2: return 0   # negative
    elif rating == 3: return 1 # neutral
    else: return 2             # positive

def get_helpfulness_label(helpful: List[int]) -> int:
    # helpful[0] = number of users who found it helpful; helpful[1] = total votes
    return 1 if helpful[0] > 0 else 0

class ReviewDataset(Dataset):
    def __init__(self, data, tokenizer, max_len):
        self.data = data
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        item = self.data[idx]
        ids = self.tokenizer.encode(item["reviewText"], max_len=self.max_len)
        length = len(ids)
        # Padding
        padded = ids + [self.tokenizer.word2idx["<PAD>"]] * (self.max_len - len(ids))
        sentiment = get_sentiment_label(item["overall"])
        helpful = get_helpfulness_label(item["helpful"])
        return (
            torch.tensor(padded, dtype=torch.long),
            torch.tensor(length, dtype=torch.long),
            torch.tensor(sentiment, dtype=torch.long),
            torch.tensor(helpful, dtype=torch.float)  # binary cross-entropy expects float
        )

train_dataset = ReviewDataset(train_data, tokenizer, MAX_SEQ_LEN)
val_dataset = ReviewDataset(val_data, tokenizer, MAX_SEQ_LEN)
test_dataset = ReviewDataset(test_data, tokenizer, MAX_SEQ_LEN)

BATCH_SIZE = 64
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)
```

# 3. Multi-Head Attention (from scratch)

```python
class MultiHeadAttention(nn.Module):
    def __init__(self, d_model: int, n_heads: int, dropout: float = 0.1):
        super().__init__()
        assert d_model % n_heads == 0
        self.d_model = d_model
        self.n_heads = n_heads
        self.d_k = d_model // n_heads

        self.W_q = nn.Linear(d_model, d_model)
        self.W_k = nn.Linear(d_model, d_model)
        self.W_v = nn.Linear(d_model, d_model)
        self.W_o = nn.Linear(d_model, d_model)
        self.dropout = nn.Dropout(dropout)

    def forward(self, query, key, value, mask=None):
        """
        query, key, value: (batch, seq_len, d_model)
        mask: (batch, 1, seq_len, seq_len) or (batch, seq_len, seq_len)
        """
        B, T, _ = query.shape

        # Linear projections and reshape for multi-head
        Q = self.W_q(query).view(B, T, self.n_heads, self.d_k).transpose(1, 2)  # (B, h, T, d_k)
        K = self.W_k(key).view(B, T, self.n_heads, self.d_k).transpose(1, 2)
        V = self.W_v(value).view(B, T, self.n_heads, self.d_k).transpose(1, 2)

        # Scaled dot-product attention
        attn_scores = torch.matmul(Q, K.transpose(-2, -1)) / math.sqrt(self.d_k)
        if mask is not None:
            # mask shape must be broadcastable to attn_scores
            attn_scores = attn_scores.masked_fill(mask == 0, float('-inf'))
        attn_weights = F.softmax(attn_scores, dim=-1)
        attn_weights = self.dropout(attn_weights)
        output = torch.matmul(attn_weights, V)  # (B, h, T, d_k)

        # Concat heads and project
        output = output.transpose(1, 2).contiguous().view(B, T, self.d_model)
        return self.W_o(output), attn_weights
```

# 4. Transformer Encoder Block

Each block consists of multi-head self-attention + residual connection + layer norm, followed by a feed‑forward network + residual + layer norm.

```python
class EncoderBlock(nn.Module):
    def __init__(self, d_model: int, n_heads: int, d_ff: int, dropout: float = 0.1):
        super().__init__()
        self.attn = MultiHeadAttention(d_model, n_heads, dropout)
        self.norm1 = nn.LayerNorm(d_model)
        self.norm2 = nn.LayerNorm(d_model)
        self.ffn = nn.Sequential(
            nn.Linear(d_model, d_ff),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(d_ff, d_model)
        )
        self.dropout = nn.Dropout(dropout)

    def forward(self, x, mask=None):
        # Self-attention
        attn_out, _ = self.attn(x, x, x, mask)
        x = self.norm1(x + self.dropout(attn_out))
        # Feed-forward
        ffn_out = self.ffn(x)
        x = self.norm2(x + self.dropout(ffn_out))
        return x
```

# 5. Encoder-Only Model (Part A)

The model uses an embedding layer (token + positional) followed by a stack of encoder blocks. A mean pooling (over non‑padding tokens) produces a fixed‑dimensional review vector. Two separate classification heads predict sentiment (3‑way) and helpfulness (binary)

```python
class PositionalEncoding(nn.Module):
    def __init__(self, d_model: int, max_len: int = 5000):
        super().__init__()
        pe = torch.zeros(max_len, d_model)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-math.log(10000.0) / d_model))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0)  # (1, max_len, d_model)
        self.register_buffer('pe', pe)

    def forward(self, x):
        # x: (batch, seq_len, d_model)
        return x + self.pe[:, :x.size(1), :]

class EncoderModel(nn.Module):
    def __init__(self, vocab_size: int, d_model: int, n_heads: int, n_layers: int,
                 d_ff: int, max_len: int, num_sent_classes: int, dropout: float = 0.1):
        super().__init__()
        self.token_embed = nn.Embedding(vocab_size, d_model, padding_idx=tokenizer.word2idx["<PAD>"])
        self.pos_encode = PositionalEncoding(d_model, max_len)
        self.dropout = nn.Dropout(dropout)
        self.encoder_blocks = nn.ModuleList([
            EncoderBlock(d_model, n_heads, d_ff, dropout) for _ in range(n_layers)
        ])
        self.sent_classifier = nn.Linear(d_model, num_sent_classes)
        self.help_classifier = nn.Linear(d_model, 1)  # binary logit

    def forward(self, input_ids, lengths):
        # input_ids: (B, T)
        B, T = input_ids.shape
        mask = self.get_pad_mask(input_ids)  # (B, T) -> (B, 1, 1, T)
        x = self.token_embed(input_ids) * math.sqrt(self.token_embed.embedding_dim)
        x = self.pos_encode(x)
        x = self.dropout(x)
        for block in self.encoder_blocks:
            x = block(x, mask)

        # Mean pooling over non-padded positions for fixed-size representation
        # Build a mask: True for real tokens
        non_pad = (input_ids != tokenizer.word2idx["<PAD>"]).unsqueeze(-1).float()  # (B, T, 1)
        pooled = (x * non_pad).sum(dim=1) / (non_pad.sum(dim=1) + 1e-8)  # (B, d_model)

        sent_logits = self.sent_classifier(pooled)
        help_logits = self.help_classifier(pooled).squeeze(-1)  # (B,)
        return sent_logits, help_logits, pooled

    def get_pad_mask(self, input_ids):
        # Create mask (B, 1, 1, T) where padding positions are 0
        pad_mask = (input_ids != tokenizer.word2idx["<PAD>"]).unsqueeze(1).unsqueeze(2)
        return pad_mask
```

Now we define training: sentiment loss (CrossEntropy), helpfulness loss (BCEWithLogits), combined sum. We’ll use AdamW with a linear warmup scheduler.

```python
# ---- Hyperparameters (feel free to experiment) ----
D_MODEL = 256
N_HEADS = 8
N_LAYERS = 4
D_FF = 512
DROPOUT = 0.2
LEARNING_RATE = 1e-4
EPOCHS = 10

vocab_size = tokenizer.vocab_size
encoder = EncoderModel(vocab_size, D_MODEL, N_HEADS, N_LAYERS, D_FF, MAX_SEQ_LEN, 3, DROPOUT).to(device)

criterion_sent = nn.CrossEntropyLoss()
criterion_help = nn.BCEWithLogitsLoss()
optimizer = torch.optim.AdamW(encoder.parameters(), lr=LEARNING_RATE, weight_decay=1e-5)
scheduler = torch.optim.lr_scheduler.OneCycleLR(optimizer, max_lr=LEARNING_RATE,
                                                steps_per_epoch=len(train_loader), epochs=EPOCHS)

def train_encoder(epoch):
    encoder.train()
    total_loss, sent_correct, help_correct, total = 0, 0, 0, 0
    loop = tqdm(train_loader, desc=f"Epoch {epoch+1}")
    for input_ids, lengths, sent_labels, help_labels in loop:
        input_ids, sent_labels, help_labels = input_ids.to(device), sent_labels.to(device), help_labels.to(device)
        optimizer.zero_grad()
        sent_logits, help_logits, _ = encoder(input_ids, lengths)
        loss_sent = criterion_sent(sent_logits, sent_labels)
        loss_help = criterion_help(help_logits, help_labels)
        loss = loss_sent + loss_help
        loss.backward()
        torch.nn.utils.clip_grad_norm_(encoder.parameters(), 1.0)
        optimizer.step()
        scheduler.step()

        total_loss += loss.item()
        sent_correct += (sent_logits.argmax(dim=1) == sent_labels).sum().item()
        help_preds = (torch.sigmoid(help_logits) > 0.5).long()
        help_correct += (help_preds == help_labels.long()).sum().item()
        total += sent_labels.size(0)
        loop.set_postfix(loss=loss.item(), sent_acc=sent_correct/total, help_acc=help_correct/total)
    return total_loss/len(train_loader), sent_correct/total, help_correct/total

def eval_encoder(loader):
    encoder.eval()
    sent_correct, help_correct, total = 0, 0, 0
    with torch.no_grad():
        for input_ids, lengths, sent_labels, help_labels in loader:
            input_ids, sent_labels, help_labels = input_ids.to(device), sent_labels.to(device), help_labels.to(device)
            sent_logits, help_logits, _ = encoder(input_ids, lengths)
            sent_correct += (sent_logits.argmax(dim=1) == sent_labels).sum().item()
            help_preds = (torch.sigmoid(help_logits) > 0.5).long()
            help_correct += (help_preds == help_labels.long()).sum().item()
            total += sent_labels.size(0)
    return sent_correct/total, help_correct/total

# Training loop
train_losses = []
val_sent_acc = []
val_help_acc = []
best_val_acc = 0.0
for epoch in range(EPOCHS):
    avg_loss, train_sent_acc, train_help_acc = train_encoder(epoch)
    train_losses.append(avg_loss)
    vs_acc, vh_acc = eval_encoder(val_loader)
    val_sent_acc.append(vs_acc)
    val_help_acc.append(vh_acc)
    print(f"Epoch {epoch+1}: Train Loss {avg_loss:.4f} | Val Sent Acc {vs_acc:.4f} | Val Help Acc {vh_acc:.4f}")
    # Save best model
    if vs_acc > best_val_acc:
        best_val_acc = vs_acc
        torch.save(encoder.state_dict(), "models/best_encoder.pt")
        print("  -> saved best encoder")

# Plot learning curves
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(train_losses, label='Train Loss')
plt.title("Training Loss")
plt.xlabel("Epoch")
plt.legend()
plt.subplot(1,2,2)
plt.plot(val_sent_acc, label='Sentiment Acc')
plt.plot(val_help_acc, label='Helpfulness Acc')
plt.title("Validation Accuracy")
plt.xlabel("Epoch")
plt.legend()
plt.tight_layout()
plt.savefig("results/encoder_curves.png")
plt.show()

# Final test evaluation with best encoder
encoder.load_state_dict(torch.load("models/best_encoder.pt", map_location=device))
test_sent_acc, test_help_acc = eval_encoder(test_loader)
print(f"Test Sentiment Accuracy: {test_sent_acc:.4f}")
print(f"Test Helpfulness Accuracy: {test_help_acc:.4f}")
```

Now we save the reference embeddings of the training set for retrieval.

```python
def save_embeddings(model, dataset, filename="results/train_embeddings.pt"):
    model.eval()
    all_embeds = []
    all_texts = []
    with torch.no_grad():
        for idx in range(len(dataset)):
            input_ids, lengths, _, _ = dataset[idx]
            input_ids = input_ids.unsqueeze(0).to(device)
            _, _, embed = model(input_ids, lengths.unsqueeze(0))
            all_embeds.append(embed.squeeze(0).cpu())
            all_texts.append(train_data[idx]["reviewText"])
    all_embeds = torch.stack(all_embeds)
    torch.save({"embeddings": all_embeds, "texts": all_texts}, filename)
    print(f"Saved {len(all_embeds)} embeddings.")

save_embeddings(encoder, train_dataset)
```

# 6. Retrieval Module (Part B)

We load the saved embeddings, build a simple index, and retrieve top‑k similar reviews via cosine similarity. Query embeddings come from the encoder for test reviews.

```python
class RetrievalModule:
    def __init__(self, embeddings_file: str):
        data = torch.load(embeddings_file, map_location="cpu")
        self.embeddings = data["embeddings"]          # (N, D)
        self.texts = data["texts"]
        # Normalize for cosine similarity
        self.norm_emb = F.normalize(self.embeddings, p=2, dim=1)

    def retrieve(self, query_embed: torch.Tensor, k: int = 5) -> List[Tuple[str, float]]:
        # query_embed: (D,)
        q_norm = F.normalize(query_embed.unsqueeze(0), p=2, dim=1)
        sims = torch.mm(q_norm, self.norm_emb.T).squeeze(0)  # (N,)
        top_sims, top_ids = torch.topk(sims, k)
        results = [(self.texts[i], sim.item()) for i, sim in zip(top_ids, top_sims)]
        return results

# Build retrieval index from training embeddings
retriever = RetrievalModule("results/train_embeddings.pt")

# Example retrieval for a test review
test_sample = test_data[0]
test_ids, test_len, _, _ = test_dataset[0]
test_ids = test_ids.unsqueeze(0).to(device)
test_len = test_len.unsqueeze(0).to(device)
with torch.no_grad():
    _, _, query_emb = encoder(test_ids, test_len)
retrieved = retriever.retrieve(query_emb.squeeze(0), k=3)
print("Original review:", test_sample["reviewText"][:200], "...")
print("\nTop retrieved reviews:")
for i, (txt, sim) in enumerate(retrieved):
    print(f"{i+1}: (sim={sim:.3f}) {txt[:100]}...")
```

# 7. Decoder-Only Transformer for Explanation Generation (Part C)

The decoder generates the summary field conditioned on:

The review text

Predicted sentiment (converted to a word token)

Predicted helpfulness (a word token)

The top‑k retrieved reviews (concatenated)

We build a prompt structure using special tokens and train a causal language model on the full concatenated sequence, computing loss only on the summary part.

7.1 Data Construction
First, we augment the dataset with prompts and target summaries. For each sample, we simulate retrieval using the encoder (in training we can cheat by using the same training embeddings, but for validation/test we must use true retrieval from the training set). To respect Part B integration, we actually run the retrieval module for every training/val/test instance (using the fixed training embeddings). This is a bit slow but possible. To avoid recomputation, we save retrieved contexts.

Given limited time in this answer, we pre‑compute retrieved contexts for all splits using the trained encoder. We’ll store the top‑k retrieved review texts.

```python
K_RETRIEVED = 3

def build_retrieved_context(dataset, encoder, retriever, k=K_RETRIEVED):
    contexts = []
    encoder.eval()
    with torch.no_grad():
        for idx in range(len(dataset)):
            input_ids, lengths, _, _ = dataset[idx]
            input_ids = input_ids.unsqueeze(0).to(device)
            _, _, embed = encoder(input_ids, lengths.unsqueeze(0))
            retrieved = retriever.retrieve(embed.squeeze(0), k=k)
            # Concatenate the retrieved reviews with a separator
            retrieved_text = " [SEP] ".join([txt for txt, _ in retrieved])
            contexts.append(retrieved_text)
    return contexts

# This may take a while; use a subset for speed but here we do all
print("Building retrieved contexts for training set...")
train_contexts = build_retrieved_context(train_dataset, encoder, retriever)
print("Building retrieved contexts for validation set...")
val_contexts = build_retrieved_context(val_dataset, encoder, retriever)
print("Building retrieved contexts for test set...")
test_contexts = build_retrieved_context(test_dataset, encoder, retriever)
```

Now we format prompts. Sentiment labels become words: "negative", "neutral", "positive". Helpfulness: "helpful" or "not helpful". The target summary is prepended with <EXP>.

Template:

text
<REVIEW> review text <SENT> sentiment <HELP> helpfulness <RETR> retrieved context <SEP> summary <EOS>
But for causal decoder training, we’ll feed the entire sequence up to <EOS> and compute loss only from the summary part onward.

We’ll encode everything with the same tokenizer, using special tokens.

```python
SENT_WORDS = {0: "negative", 1: "neutral", 2: "positive"}
HELP_WORDS = {0: "not helpful", 1: "helpful"}

class ExplanationDataset(Dataset):
    def __init__(self, data, contexts, dataset_split, tokenizer, max_len=512):
        self.data = data
        self.contexts = contexts
        self.tokenizer = tokenizer
        self.max_len = max_len
        # Precompute encoded prompts and targets
        self.encodings = []
        for i, item in enumerate(data):
            sent = get_sentiment_label(item["overall"])
            help_ = get_helpfulness_label(item["helpful"])
            prompt = (
                f"<REVIEW> {item['reviewText']} <SENT> {SENT_WORDS[sent]} "
                f"<HELP> {HELP_WORDS[help_]} <RETR> {contexts[i]} <SEP>"
            )
            target = f" <EXP> {item.get('summary', '')} <EOS>"
            # Combine and tokenize
            full_text = prompt + target
            ids = self.tokenizer.encode(full_text, max_len=self.max_len, add_special=False)
            # Create attention mask and labels
            # For causal LM, we shift right; we'll handle in collate
            self.encodings.append(ids)
    
    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return torch.tensor(self.encodings[idx], dtype=torch.long)

# Note: We need a custom collate to pad sequences and create causal masks.
def collate_fn(batch):
    max_len = max(len(seq) for seq in batch)
    padded = torch.zeros(len(batch), max_len, dtype=torch.long).fill_(tokenizer.word2idx["<PAD>"])
    for i, seq in enumerate(batch):
        padded[i, :len(seq)] = seq
    return padded

exp_train = ExplanationDataset(train_data, train_contexts, train_dataset, tokenizer, max_len=512)
exp_val = ExplanationDataset(val_data, val_contexts, val_dataset, tokenizer, max_len=512)
exp_test = ExplanationDataset(test_data, test_contexts, test_dataset, tokenizer, max_len=512)

exp_train_loader = DataLoader(exp_train, batch_size=32, shuffle=True, collate_fn=collate_fn)
exp_val_loader = DataLoader(exp_val, batch_size=32, shuffle=False, collate_fn=collate_fn)
exp_test_loader = DataLoader(exp_test, batch_size=32, shuffle=False, collate_fn=collate_fn)
```

# 7.2 Decoder-Only Transformer

We implement a decoder block with causal self‑attention (masked) and no cross‑attention. The model predicts the next token, conditioned on previous tokens.

```python
class DecoderBlock(nn.Module):
    def __init__(self, d_model: int, n_heads: int, d_ff: int, dropout: float = 0.1):
        super().__init__()
        self.attn = MultiHeadAttention(d_model, n_heads, dropout)
        self.norm1 = nn.LayerNorm(d_model)
        self.norm2 = nn.LayerNorm(d_model)
        self.ffn = nn.Sequential(
            nn.Linear(d_model, d_ff),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(d_ff, d_model)
        )
        self.dropout = nn.Dropout(dropout)

    def forward(self, x, causal_mask):
        # x: (B, T, d_model)
        # causal_mask: (T, T) with False for future positions
        attn_out, _ = self.attn(x, x, x, mask=causal_mask.unsqueeze(0).unsqueeze(0))
        x = self.norm1(x + self.dropout(attn_out))
        ffn_out = self.ffn(x)
        x = self.norm2(x + self.dropout(ffn_out))
        return x

class DecoderModel(nn.Module):
    def __init__(self, vocab_size: int, d_model: int, n_heads: int, n_layers: int,
                 d_ff: int, max_len: int, dropout: float = 0.1):
        super().__init__()
        self.token_embed = nn.Embedding(vocab_size, d_model, padding_idx=tokenizer.word2idx["<PAD>"])
        self.pos_encode = PositionalEncoding(d_model, max_len)
        self.dropout = nn.Dropout(dropout)
        self.dec_blocks = nn.ModuleList([
            DecoderBlock(d_model, n_heads, d_ff, dropout) for _ in range(n_layers)
        ])
        self.ln_final = nn.LayerNorm(d_model)
        self.output = nn.Linear(d_model, vocab_size)

    def forward(self, input_ids):
        """
        input_ids: (B, T) padded
        Returns logits over vocabulary for each position.
        """
        B, T = input_ids.shape
        # Causal mask: upper triangle set to 0 (future tokens)
        causal_mask = torch.triu(torch.ones(T, T, device=device), diagonal=1).bool()
        causal_mask = ~causal_mask  # True positions are allowed (not masked)

        x = self.token_embed(input_ids) * math.sqrt(self.token_embed.embedding_dim)
        x = self.pos_encode(x)
        x = self.dropout(x)
        for block in self.dec_blocks:
            x = block(x, causal_mask)
        x = self.ln_final(x)
        logits = self.output(x)  # (B, T, vocab_size)
        return logits

    def generate(self, input_ids, max_new_tokens=50, temperature=1.0):
        self.eval()
        with torch.no_grad():
            for _ in range(max_new_tokens):
                logits = self.forward(input_ids)          # (1, T, V)
                logits = logits[:, -1, :] / temperature    # last token
                probs = F.softmax(logits, dim=-1)
                next_token = torch.multinomial(probs, num_samples=1)
                if next_token.item() == tokenizer.word2idx["<EOS>"]:
                    break
                input_ids = torch.cat([input_ids, next_token], dim=1)
        return input_ids
```

# 7.3 Training the Decoder

We compute cross‑entropy loss only on tokens that belong to the explanation (i.e., after the <SEP> token). We also create a mask to ignore padding and the prompt part.

```python
# Hyperparameters for decoder
DEC_D_MODEL = 256
DEC_N_HEADS = 8
DEC_N_LAYERS = 4
DEC_D_FF = 512
DEC_DROPOUT = 0.2
DEC_LR = 1e-4
DEC_EPOCHS = 10

decoder = DecoderModel(vocab_size, DEC_D_MODEL, DEC_N_HEADS, DEC_N_LAYERS, DEC_D_FF, MAX_SEQ_LEN, DEC_DROPOUT).to(device)

optimizer_dec = torch.optim.AdamW(decoder.parameters(), lr=DEC_LR, weight_decay=1e-5)
scheduler_dec = torch.optim.lr_scheduler.OneCycleLR(optimizer_dec, max_lr=DEC_LR,
                                                    steps_per_epoch=len(exp_train_loader), epochs=DEC_EPOCHS)

SEP_ID = tokenizer.word2idx["<SEP>"]
PAD_ID = tokenizer.word2idx["<PAD>"]

def train_decoder(epoch):
    decoder.train()
    total_loss = 0
    loop = tqdm(exp_train_loader, desc=f"Dec epoch {epoch+1}")
    for batch in loop:
        batch = batch.to(device)
        B, T = batch.shape
        # Input is batch[:, :-1], target is batch[:, 1:] (next token prediction)
        input_ids = batch[:, :-1]
        targets = batch[:, 1:]

        logits = decoder(input_ids)   # (B, T-1, V)
        # Create mask: only compute loss on tokens after <SEP> and not padding
        # Find position of <SEP> in each sequence (batch[:, 1:] alignment)
        sep_positions = (targets == SEP_ID).long().argmax(dim=1)  # first occurrence
        mask = torch.zeros_like(targets, dtype=torch.bool)
        for b in range(B):
            start = sep_positions[b] + 1  # start after <SEP>
            if start < T-1:
                mask[b, start:] = True
        # Also ignore padding
        mask = mask & (targets != PAD_ID)

        loss = F.cross_entropy(logits.permute(0,2,1), targets, reduction='none')
        loss = (loss * mask.float()).sum() / mask.sum()

        optimizer_dec.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(decoder.parameters(), 1.0)
        optimizer_dec.step()
        scheduler_dec.step()
        total_loss += loss.item()
        loop.set_postfix(loss=loss.item())
    return total_loss/len(exp_train_loader)

def eval_decoder_perplexity(loader):
    decoder.eval()
    total_loss, total_tokens = 0, 0
    with torch.no_grad():
        for batch in loader:
            batch = batch.to(device)
            B, T = batch.shape
            input_ids = batch[:, :-1]
            targets = batch[:, 1:]
            logits = decoder(input_ids)
            sep_positions = (targets == SEP_ID).long().argmax(dim=1)
            mask = torch.zeros_like(targets, dtype=torch.bool)
            for b in range(B):
                start = sep_positions[b] + 1
                if start < T-1:
                    mask[b, start:] = True
            mask = mask & (targets != PAD_ID)
            loss = F.cross_entropy(logits.permute(0,2,1), targets, reduction='none')
            loss = (loss * mask.float()).sum()
            total_loss += loss.item()
            total_tokens += mask.sum().item()
    avg_loss = total_loss / total_tokens
    return math.exp(avg_loss)  # perplexity

best_ppl = float('inf')
for epoch in range(DEC_EPOCHS):
    train_loss = train_decoder(epoch)
    val_ppl = eval_decoder_perplexity(exp_val_loader)
    print(f"Epoch {epoch+1}: Train Loss {train_loss:.4f} | Val Perplexity {val_ppl:.2f}")
    if val_ppl < best_ppl:
        best_ppl = val_ppl
        torch.save(decoder.state_dict(), "models/best_decoder.pt")
        print("  -> saved best decoder")

# Test perplexity
decoder.load_state_dict(torch.load("models/best_decoder.pt", map_location=device))
test_ppl = eval_decoder_perplexity(exp_test_loader)
print(f"Test Perplexity: {test_ppl:.2f}")
```

# 7.4 Qualitative Evaluation & Generation

We generate explanations for a few test samples and compare with the baseline (no retrieved context).

```python
def generate_explanation(review, sentiment_label, helpful_label, context, model, tokenizer, max_len=512, gen_len=30):
    prompt = (
        f"<REVIEW> {review} <SENT> {SENT_WORDS[sentiment_label]} "
        f"<HELP> {HELP_WORDS[helpful_label]} <RETR> {context} <SEP>"
    )
    input_ids = torch.tensor(tokenizer.encode(prompt, max_len=max_len, add_special=False),
                             dtype=torch.long).unsqueeze(0).to(device)
    output_ids = model.generate(input_ids, max_new_tokens=gen_len)
    # Decode only the new tokens (after the prompt length)
    full_text = tokenizer.decode(output_ids.squeeze().tolist())
    # Extract the part after <SEP> (or the last generated summary)
    parts = full_text.split("<SEP>")
    if len(parts) > 1:
        explanation = parts[1].strip()
    else:
        explanation = full_text
    return explanation

# Select a few test examples for demonstration
num_examples = 5
for i in range(num_examples):
    item = test_data[i]
    sent = get_sentiment_label(item["overall"])
    hlp = get_helpfulness_label(item["helpful"])
    ctx = test_contexts[i]
    gen_exp = generate_explanation(item["reviewText"], sent, hlp, ctx, decoder, tokenizer)
    print(f"--- Example {i+1} ---")
    print(f"Review: {item['reviewText'][:100]}...")
    print(f"Generated explanation: {gen_exp}")
    print()

# Ablation: baseline without retrieval (empty context)
def baseline_explanation(review, sent, hlp, model, tokenizer):
    return generate_explanation(review, sent, hlp, "", model, tokenizer)

print("\n===== Ablation: No retrieval =====")
for i in range(num_examples):
    item = test_data[i]
    sent = get_sentiment_label(item["overall"])
    hlp = get_helpfulness_label(item["helpful"])
    gen_no_ret = baseline_explanation(item["reviewText"], sent, hlp, decoder, tokenizer)
    print(f"Example {i+1}: {gen_no_ret}")
```
